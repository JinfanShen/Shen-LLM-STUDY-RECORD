# 模型调用

```python
seq = "那就这样吧"
tokenizer = AutoTokenizer.from_pretrained('hfl/rbt3')
inputs = tokenizer(seq, return_tensors='pt')  #返回tensor类型的id
print(inputs)
```

# 不带Model Head的模型调用

```python
model = AutoModel.from_pretrained('hfl/rbt3') 
```

```python
output = model(**inputs)
# print(output)
print(output.last_hidden_state.shape)
```
